```python
from pathlib import Path
from basic_pitch.inference import predict_and_save
from basic_pitch import ICASSP_2022_MODEL_PATH
from multiprocessing import Pool
import tqdm as notebook_tqdm
import os
import json
from concurrent.futures import ThreadPoolExecutor, as_completed
import threading
```

```python
source_folder = Path(r"E:\youtube_audio_first_24285")
destination_folder = Path(r"E:\youtube_audio_midi")
progress_file = Path(r"E:\youtube_audio_midi\progress.json")
destination_folder.mkdir(parents=True, exist_ok=True)

# Thread-safe lock for progress updates
progress_lock = threading.Lock()

def load_progress():
    if progress_file.exists():
        with open(progress_file) as f:
            return set(json.load(f))
    return set()

def save_progress(completed):
    with open(progress_file, 'w') as f:
        json.dump(list(completed), f)

def convert_file(audio_file, destination_folder, completed, pbar):
    """Convert a single audio file and update progress."""
    try:
        input_path = str(audio_file)
        predict_and_save([input_path], str(destination_folder), True, False, False, False, ICASSP_2022_MODEL_PATH)
        
        with progress_lock:
            completed.add(audio_file.name)
            save_progress(completed)
            pbar.update(1)
        
        return audio_file.name, None  # (filename, error)
    
    except Exception as e:
        return audio_file.name, e

# --- Main ---
audio_extensions = {'.wav', '.mp3', '.flac', '.m4a', '.ogg'}
all_files = [f for f in source_folder.iterdir() if f.suffix.lower() in audio_extensions]

completed = load_progress()
remaining = [f for f in all_files if f.name not in completed]

print(f"Total: {len(all_files)} | Done: {len(completed)} | Remaining: {len(remaining)}\n")

NUM_WORKERS = 12  # Tune based on CPU/GPU — try 2–4 for GPU-bound models

if remaining:
    try:
        with notebook_tqdm.tqdm(total=len(remaining), desc="Converting") as pbar:
            with ThreadPoolExecutor(max_workers=NUM_WORKERS) as executor:
                futures = {
                    executor.submit(convert_file, f, destination_folder, completed, pbar): f
                    for f in remaining
                }
                for future in as_completed(futures):
                    filename, error = future.result()
                    if error:
                        print(f"\n✗ Failed: {filename} — {error}")

        print(f"\n✓ All done! {len(completed)} files converted")

    except KeyboardInterrupt:
        print("\n⏸ Paused. Progress saved — run again to resume.")
        with progress_lock:
            save_progress(completed)
else:
    print("All files already converted!")
```

```python
source_folder = Path(r"E:\youtube_audio")
destination_folder = Path(r"E:\youtube_audio_midi")
progress_file = Path(r"E:\youtube_audio_midi\progress.json")

destination_folder.mkdir(parents=True, exist_ok=True)

def load_progress():
    """Load already converted files"""
    if progress_file.exists():
        with open(progress_file) as f:
            return set(json.load(f))
    return set()

def save_progress(completed):
    """Save progress to file"""
    with open(progress_file, 'w') as f:
        json.dump(list(completed), f)

def convert_audio(audio_file):
    """Convert single audio file to MIDI"""
    try:
        input_path = str(audio_file)
        predict_and_save([input_path], str(destination_folder), True, False, False, False, ICASSP_2022_MODEL_PATH)
        return audio_file.name, input_path, None
    except Exception as e:
        return None, str(source_folder / audio_file), str(e)

# Get all audio files
audio_extensions = {'.wav', '.mp3', '.flac', '.m4a', '.ogg'}
all_files = [f for f in source_folder.iterdir() if f.suffix.lower() in audio_extensions]

# Load already converted files
completed = load_progress()
remaining = [f for f in all_files if f.name not in completed]

print(f"Total: {len(all_files)} | Already done: {len(completed)} | Remaining: {len(remaining)}\n")

if remaining:
    try:
        with Pool(os.cpu_count()) as pool:
            for result in notebook_tqdm.tqdm(pool.imap_unordered(convert_audio, notebook_tqdm.tqdm(remaining, desc="Converting")), total=len(remaining), desc="Converting"):
                if result:
                    completed.add(result)
                    save_progress(completed)
        
        print(f"\n✓ All done! {len(completed)} files converted")
        
    except KeyboardInterrupt:
        print("\n⏸ Paused. Run again to resume from where you left off.")
        save_progress(completed)
else:
    print("All files already converted!")
```
